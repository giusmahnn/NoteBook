```python
# I import the necessary modules needed for wrangling the data

import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDA0101ENSkillsNetwork20235326-2022-01-01'
c_c = pd.read_csv(url, header=None)
```

```python
c_c.head()
```

```python
# I assinged the name of headers and loaded it into the dataframe using the df.columns

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
c_c.columns = headers
```

```python
# I used the .replace to put Nan in place of the ? mark

c_c.replace('?', np.nan, inplace=True)
c_c.head()
```

```python
# I used the code below to get the total number of missing values in each column
c_c.isna().sum()
```

```python
#Replacing certain columns by mean

avg_normalized_loss = round(c_c["normalized-losses"].astype("float").mean(axis = 0), 2)
print('Average normalized loss = ', avg_normalized_loss)
avg_bore=round(c_c['bore'].astype('float').mean(axis=0), 2)
print('Average bore = ', avg_bore)
avg_stroke = round(c_c["stroke"].astype("float").mean(axis = 0), 2)
print('Average stroke = ', avg_stroke)
avg_horsepower = round(c_c['horsepower'].astype('float').mean(axis=0), 2)
print('Average horsepower = ', avg_horsepower)
avg_peakrpm=round(c_c['peak-rpm'].astype('float').mean(axis=0), 2)
print('Average peak-rpm = ', avg_peakrpm)
```

```python
c_c.describe(include='all')
```

```python
# Replace missing values with the mean
c_c["normalized-losses"].replace(np.nan, avg_normalized_loss, inplace=True)
c_c["bore"].replace(np.nan, avg_bore, inplace=True)
c_c["stroke"].replace(np.nan, avg_stroke, inplace=True)
c_c["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
c_c["peak-rpm"].replace(np.nan, avg_peakrpm, inplace=True)

# Replace door with the most occuring frequency
c_c["num-of-doors"].replace(np.nan, 'four', inplace=True)
c_c.head()
```

```python
# I dropped the missing values in the prices row
c_c.dropna(subset=['price'], inplace=True)

# Reset the index of the dataset
c_c.reset_index(drop=True, inplace=True)
```

```python
# I corrected the data types of the columns
c_c[["bore", "stroke", "price", "peak-rpm"]] = c_c[["bore", "stroke", "price", "peak-rpm"]].astype("float")
c_c[["normalized-losses"]] = c_c[["normalized-losses"]].astype("int64")

c_c.dtypes
```

NORMALIZATION AND BINNING OF DATA(Horsepower and Engine-size)

```python
c_c['length'] = c_c['length']/c_c['length'].max()
c_c['width'] = c_c['width']/c_c['width'].max()
c_c['height'] = c_c['height']/c_c['height'].max()
c_c[['length', 'width', 'height']].head()
```

```python
# converted the data types from object to int
c_c['horsepower'] = c_c['horsepower'].astype(int, copy=True)

# I used the numpy built-in function to group the column into 3 groups
bins = np.linspace(min(c_c['horsepower']), max(c_c['horsepower']), 4)

#create group names
g_names = ['low', 'medium', 'high']

# create a new column for the binned data
c_c['horsepower-binned'] = pd.cut(c_c['horsepower'], bins,labels=g_names, include_lowest=True)
c_c[['horsepower', 'horsepower-binned']].head(10)
```

```python
# we get the values for each group
c_c['horsepower-binned'].value_counts()
```

```python
# I visualized the data with a barchart using matplotlib
plt.bar(g_names, c_c["horsepower-binned"].value_counts())

# name for the x-axis
plt.xlabel("Horsepower")

# name for y-axis
plt.ylabel("Count")

# Title of the barchart
plt.title("Horsepower bins")
```

```python
# I used the numpy built-in function to group the column into 3 groups
bins = np.linspace(min(c_c['engine-size']), max(c_c['engine-size']), 4)

#create group names
g_names = ['low', 'medium', 'high']

# create a new column for the binned data
c_c['engine-size-binned'] = pd.cut(c_c['engine-size'], bins,labels=g_names, include_lowest=True)
c_c[['engine-size', 'engine-size-binned']].head(10)
```

```python
# i got the value of the binned engine-size(values in each group)
c_c['engine-size-binned'].value_counts()
```

```python
plt.bar(g_names, c_c["engine-size-binned"].value_counts())

plt.xlabel("Engine-size")
plt.ylabel("Count")
plt.title("Engine-Size bins")
```

CREATING DUMMIES OF DATA(Fuel-type and Aspirations)

```python
dum_fuel = pd.get_dummies(c_c['fuel-type'])
dum_fuel.rename(columns={"gas": "fuel-gas", "diesel":"fuel-diesel"}, inplace=True)
c_c = pd.concat([c_c, dum_fuel], axis = 1)
c_c.drop('fuel-type', axis=1, inplace=True)
dum_fuel.head()
```

```python
dum_asp = pd.get_dummies(c_c['aspiration'])
dum_asp.rename(columns={"std": "aspiration-std", "turbo":"aspiration-turbo"}, inplace=True)
c_c = pd.concat([c_c, dum_asp], axis = 1)
c_c.drop('aspiration', axis=1, inplace=True)
dum_asp.head()
```
