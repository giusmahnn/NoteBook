```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# we passed this line of code to show all columns and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Loaded the dataset into panda
df = pd.read_csv('ChicagoPublicSchools_data.csv', low_memory=False)

# first four rows of the dataset
df.head()
```

```python
# we printed out the shape of the dataset to know how much data we are working with
print('There are',df.shape[1],'Columns and',df.shape[0],'rows in the dataset.')
```

```python
# Info about our dataset and the data type
df.info()
```

```python
df.drop(['adequate_yearly_progress_made_', 'family_involvement_score', 'family_involvement_icon',])
```

```python
# Converted the column names to lower case

df.columns = [column.replace('__','').lower() for column in df.columns]

# printed out the column to check if it worked
df.head(n=0)
```

```python
# We checked to see if there are null values
# It prints out the total amount
# The columns containing the null values

m = df.isna().sum().to_list()
r = sum(m)

if r == 0:
    print("_No NaN Values in dataframe_")
else:
    print(df.isna().values.sum(),"NaN values exists")
    print('This is a list of columns containing NaN values: \n',df.columns[df.isna().any()])
```

```python
# Replaced the null value in the link column to a 'no link available'.
df['link'] = df['link'].apply(lambda x: 'no link availabe' if pd.isna(x) else x)
df['link'].isna().sum()
```

```python
data = ['safety_score', 'environment_score', 'instruction_score', 'isat_exceeding_math', 'isat_exceeding_reading',
       'isat_value_add_math', 'isat_value_add_read']
for r in df[data]:
    df[r] = df[r].apply(lambda x: '0.00' if pd.isna(x) else x)
    df[r] = df[r].astype('float64')

df[data].isna().sum()
```

```python
cols = ['average_student_attendance', 'average_teacher_attendance', 'individualized_education_program_compliance_rate']
for c in df[cols]:
    df[c] = df[c].apply(lambda x: '0.00' if pd.isna(x) else x)
    df[c] = df[c].apply(lambda x: x[:-1] if '%' in x else x)
    df[c] = df[c].astype('float64')
new_cols = [col + '(%)' if col in cols else col for col in df.columns]
df.columns = new_cols
```
