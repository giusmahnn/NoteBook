```python
import pandas as pd
```

```python
# displaying all columns in the dataset
pd.set_option('display.max_columns', None)

#reading and displaying the dataset in pandas
df = pd.read_csv('fifa21.csv', low_memory=False)
df.head()
```

```python
# we get a little information about each column in the dataset
df.info()
```

```python
# every space in between two words are changed to underscore and the uppercase changed to lowercase
df.columns = [column.replace(' ', '_').lower() for column in df.columns]

# columns with special characters are replaced
df.rename(columns={'↓ova':'ova'}, inplace= True)
df.columns
```

```python
# we check for the sum of missing values in each column of the dataset
df.isna().sum()
```

```python
# this are the columns found to be messy and will be worked on
dirty = ['hits','loan_date_end', 'ir', 'sm', 'w/f', 'value', 'release_clause', 'wage', 
         'club', 'contract','joined','height', 'weight']

df[dirty].head()
```

```python
# I used lambda function to multiply the removed K by 1000
df['hits'] = df['hits'].apply(lambda x: pd.to_numeric(x[:-1], errors='coerce') * 1000 if 'K' in str(x) and '.' in str(x) else pd.to_numeric(x, errors='coerce'))

# The missing values are replaced with 0
df['hits'] = df['hits'].apply(lambda x: x if not pd.isnull(x) else "0")

# we converted to int64
df['hits'] = df['hits'].astype('int64')
```

```python
# we created a new column name 'loan_status'
# missing values are replaced with 'on loan'
# non missing values are replaced with 'not on loan'
df['loan_status'] = df['loan_date_end'].apply(lambda x: 'not on loan' if pd.isnull(x) else 'on loan').tolist()

# The new column is added to our df
dirty.append('loan_status')

#The initial column is the deleted
dirty.remove('loan_date_end')

# we use this to display at random
df['loan_status'].sample(n=5, random_state=4)
```

```python
# The \n is removed from the column
df['club'] = df['club'].apply(lambda x: x.replace('\n',''))

# dispaly
df['club'].head()
```

```python
# A new column is created 
# EVery lbs value is converted to kg 
# The kg and lbs unit is removed from the column
df['weight(kg)'] = df['weight'].apply(lambda x: x[:-2] if 'kg' in x 
                                      else pd.to_numeric(x.replace('lbs', ''), errors='coerce') * 0.454 if 'lbs' in x else x)

# the new column is added to the dataset
dirty.append('weight(kg)')

# The initial column is removed from the dirty column list
dirty.remove('weight')

# display
df['weight(kg)'].head()
```

```python
# The qoutes are removed from the column
df['height'] = df['height'].apply(lambda x: x.replace('"', ""))
df['height'] = df['height'].apply(lambda x: x.replace("'", "."))

# A new column is created while values in feet and inches are converted to cm
# The cm unit is also removed
df['height(cm)'] = df['height'].apply(lambda x: x[:-2] if 'cm' in x 
                                      else pd.to_numeric(x, errors='coerce') * 30.48)

# The column is added 
dirty.append('height(cm)')

# The old column is removed
dirty.remove('height')

# display
df['height(cm)'].head()
```

```python
# I put columns having similarities in a list
money = ['value', 'wage', 'release_clause']
# I Used an iterator to iterate over each column
# every euros sign is removed
# values with M are multiplied by '1000000'
# values with K are multipled by '1000'
for r in money:
    df[r] = df[r].apply(lambda x: x[1:] if '€' in x else x)
    df[r] = df[r].apply(lambda x: pd.to_numeric(x[:-1]) * 1000000 if 'M' in x 
                        else pd.to_numeric(x[:-1]) * 1000 if 'K' in x
                        else pd.to_numeric(x, errors='coerce'))
    
    # The data type is converted to float
    df[r] = df[r].astype('float64')

# display at random
df[money].sample(n=10, random_state=4)
```

```python
# similar columns containing stars
rating = ['ir', 'sm', 'w/f']

# Using lambda and list slicing the stars are removed
for r in df[rating]:
    df[r] = df[r].apply(lambda x: x[:-1] if '★' in x else int(x))
    
# display
df[rating].head()
```

```python
# We will move on to the contract column.

# Adjusting the 'contract' column to take only useful values while ignoring '~' and other words.
df['contract_starts'] = df['contract'].apply(lambda x: x[-7:] if "On Loan" in x else x[:4].strip())

# Creating a new column called 'contract_ends' and taking values from the 'contract column' to populate it
df['contract_ends'] = df['contract'].apply(lambda x: x[:-7] if "On Loan" in x else x[:] if "Free" in x else x[-4:])

# Removing the 'contract' item from the dirty list
dirty.remove('contract')

# Adding the two new items we created to the dirty data list
dirty.extend(['contract_starts', 'contract_ends'])

# display
df[['contract_starts', 'contract_ends']].sample(n=5, random_state=4)
```

```python
# This columns each consist of the sum of 5 different columns
cols = ['movement', 'skill', 'attacking', 'mentality', 'goalkeeping', 'power', 'defending']

# I used an iterator to change every value to percentage
for r in cols:
    df[r] = df[r].apply(lambda x: round((x/500) * 100), 2)

# dispaly    
df[cols].sample(n=10, random_state=4)
```

```python
# the columns are still stored in a list to enable name change
cols = ['movement', 'skill', 'attacking', 'mentality', 'goalkeeping', 'power', 'defending']

# for every column in cols the '(%)' was added at the end of its name
new_cols = [col + '(%)' if col in cols else col for col in df.columns]

# added to the dataset
df.columns = new_cols
```

```python
# columns to be dropped from the dataset
drop_cols = ['contract', 'height', 'weight', 'loan_date_end']

# the columns are droped 
for r in drop_cols:
    df.drop([r], axis=1, inplace=True)
    
```

```python
# Checking if there are null values in the DataFrame after cleaning
r = df.isnull().sum().to_list()
m = sum(r)
if m == 0:
    print("_No NaN Values in dataframe_")
else:
    print("NaN values exists")
```
